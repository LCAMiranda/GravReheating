# Gravitational particle production

Dada a ação de um campo escalar $\varphi$ no background FRLW: 

$$
\begin{equation}
    S = -\frac{1}{2}\int d^4 x \sqrt{-g} \left(g^{\mu\nu}\partial_\mu \varphi \partial_\nu \varphi + m_\varphi ^2 \varphi^2 + \xi R\varphi ^2 \right) ,
\end{equation}
$$
podemos determinar a equação de movimento para o campo
$$ \Box \varphi - m_\varphi ^2 \varphi - \xi R\varphi = 0 $$.

Decompondo o campo em modos de Fourier, cada modo $\varphi_k (t)$ seguirá a seguinte equação de movimento:

$$\ddot \varphi_k + 3H\dot \varphi_k + \left( \frac{k^2}{a^2} +  m_\varphi ^2 + \xi R \right)\varphi_k = 0  ,$$ 

o termo de fricção presente na equação dificulta a solução, podemos eliminar o termo introduzindo um novo campo $\chi = a \varphi$, e também o tempo conforme $\eta$. Reescrevendo a equação de movimento:
$$ \chi_k^{''}+ \left[k^2 +  a^2 m_\chi ^2 + \left( \xi -\frac{1}{6}\right)a^2 R\right]\chi_k = 0  ,$$
com as devidas condições iniciais, podemos resolver a equação acima diretamente, porém estamos interessados resolver via transformação de bogoliubov.


A ação do campo $\chi$ no tempo conforme é dada por
$$S= -\frac{1}{2}\int d \eta \int d^3 x (\chi' ^2 + \omega^2 \chi) ,$$
sendo $\omega^2 = \nabla^2 + a^2 m_\chi ^2 + \frac{1}{6}a^2 R$,

expressando a equação de movimento em função do momento conjugado de $\chi$, temos que $\pi' + \omega^2 \chi = 0$. Expandindo o campo $\chi$ em modos de Fourier:

$$\chi (x) = \int \frac{d^3 k}{(2\pi)^{3/2}} e^{i \vec k \cdot \vec x} \chi_k$$

$$\chi_k = a_{\vec k}u_k + a_{-\vec k}^\dagger u_k ^*,$$

podemos também decompor $u_k$ em função dos coeficientes de Bogoliubov $A_k$ e $B_k$:

$$u_k = \frac{A_k}{\sqrt {2 \omega_k}} e^{-i \int \omega_k d\eta} + \frac{B_k}{\sqrt {2 \omega_k}} e^{i \int \omega_k d\eta},$$

para facilitar os cálculos definimos os novos coeficientes $\alpha_k \equiv A_k e^{-i \int \omega_k d\eta}$, $\beta_k \equiv B_k e^{i \int \omega_k d\eta}$.

Parametrizando o momento conjugado da mesma maneira que o campo $\chi$

$$\pi (x) = \int \frac{d^3 k}{(2\pi)^{3/2}} e^{i \vec k \cdot \vec x} \pi_k$$

$$\pi_k = a_{\vec k}v_k + a_{-\vec k}^\dagger v_k ^*,$$

$$v_k = \frac{-i \omega_k\alpha_k}{\sqrt {2 \omega_k}} + \frac{i\omega_k\beta_k}{\sqrt {2 \omega_k}},$$

para que a expressão determinada para $\chi_k$ seja solução da equação de movimento e também satisfaça a relação $\pi = \chi'$, os coeficientes devem respeitar o seguinte sistema de equações diferenciais:

$$\alpha_k '= -i\omega_k \alpha_k + \frac{\omega_k '}{2\omega_k} \beta_k$$

$$\beta_k '= i\omega_k \beta_k + \frac{\omega_k '}{2\omega_k} \alpha_k.$$

A densidade numérica de partículas criadas é dada por:

$$n = \frac{1}{2 \pi a^3}\int k^2 |\beta_k|^2 dk$$



## Escalar de Ricci

$$R(\eta^\star)/m_\phi ^2 = \frac{54\pi^{3/2}}{{(1 + \eta + \delta)^6}}    \left(1 - 3\cos\left(\frac{2}{3\sqrt{\pi}}  (1 + \eta + \delta)^3\right)\right)$$


Sendo $\delta$ o deslocamento necessário para retirar o período sem oscilação do R e $\eta^\star = \frac{\sqrt{12 \pi}}{2}m_\phi\eta$

```python
import numpy as np
import matplotlib.pyplot as plt
import cmath

t_max = 10
num_points = 1000
t = np.linspace(0, t_max, num_points)

#aqui estou utilizando delta=0.5335
R = (600/(1.5335+t)**6)*(0.5*(1 - 3*np.cos(0.37*(t+1.5335)**3)))

plt.plot(t, R, label='$R$')
plt.xlabel(r'$\eta^\star$')
plt.ylabel(r'$R$')
plt.title('Escalar de Ricci')
#plt.xlim(1,10)
#plt.ylim(-1,40)
plt.legend()
plt.show()
```

## Frequência

$$\omega/m_\phi = \sqrt{\frac{k^2}{m^2_\phi} +  a^2(\eta^\star) \frac{m_\chi ^2}{m_\phi ^2} + \left(\xi -\frac{1}{6}  \right)a^2(\eta^\star) \frac{R(\eta^\star)}{m_\phi ^2}},$$

sendo

$$a(\eta^\star) = (1 + \eta^\star)^2$$

$$
\omega'/m_\phi= \frac{1}{2\omega} \left[4  m_{\chi}^2  (1+\eta^\star)^3 + 54\pi ^{3/2} \left(\xi - \frac{1}{6}\right)\left[  9  \sin\left(\frac{2}{3\sqrt{\pi}} (1 + \eta^\star + \delta)^3\right)  \frac{2}{3\sqrt{\pi}} \left(\frac{(1+\eta^\star)^4}{(1 + \eta^\star + \delta)^4}\right) - 2  \left(1 - 3  \cos\left(\frac{2}{3\sqrt{\pi}}  (1 + \eta^\star + \delta)^3\right)\right)  \frac{\eta^\star  (1+\eta^\star)^3}{(1 + \eta^\star + \delta)^7}\right]\right]
$$


```python



# Parâmetros para cálculos
k = 10
k2 = 5
k3 = 0.1
mx = 1.5  #razão m_chi/n=m_phi
xi = 11/60

a = (t + 1) ** 2

R = (600/(1.5335+t)**6)*(0.5*(1 - 3*np.cos(0.37*(t+1.5335)**3)))

w = np.sqrt(k ** 2 + (a ** 2) * mx ** 2 + (a ** 2) * (xi - 1 / 6) * R)
w2 = np.sqrt(k2 ** 2 + (a ** 2) * mx ** 2 + (a ** 2) * (xi - 1 / 6) * R)
w3 = np.sqrt(k3 ** 2 + (a ** 2) * mx ** 2 + (a ** 2) * (xi - 1 / 6) * R)

# Plotando o gráfico
plt.plot(t, w, label='$k=10$')
plt.plot(t, w2, label='$k=5$')
plt.plot(t, w3, label='$k=0.7$')
plt.xlabel(r'$\eta^\star$')
plt.ylabel('$\omega_k / m_\phi$')
plt.title(r'Frequência $\omega_k / m_\phi , \xi = 11/60 , m_\chi/m_\phi = 1.5$')
plt.ylim(-5, 40)
plt.legend()
plt.grid()
plt.show()
wp = 0.5*(1/w)*(4*(mx**2)*((1+t)**3) + 300*(xi-1/6)*(9*np.sin(0.37*(1.5335+t)**3)*0.37*(((1+t)**4)/(1.5335 + t)**4) - 2*(1-3*np.cos(0.37*(1.5335+t)**3))*(((t*(1+t)**3)/(1.5335 + t)**7))))
wp2= 0.5*(1/w2)*(4*(mx**2)*((1+t)**3) + 300*(xi-1/6)*(9*np.sin(0.37*(1.5335+t)**3)*0.37*(((1+t)**4)/(1.5335 + t)**4) - 2*(1-3*np.cos(0.37*(1.5335+t)**3))*(((t*(1+t)**3)/(1.5335 + t)**7))))
wp3 = 0.5*(1/w3)*(4*(mx**2)*((1+t)**3) + 300*(xi-1/6)*(9*np.sin(0.37*(1.5335+t)**3)*0.37*(((1+t)**4)/(1.5335 + t)**4) - 2*(1-3*np.cos(0.37*(1.5335+t)**3))*(((t*(1+t)**3)/(1.5335 + t)**7))))
plt.plot(t, wp, label='$k=10$')
plt.plot(t, wp2, label='$k=5$')
plt.plot(t, wp3, label='$k=0.1$')
plt.xlabel(r'$\eta^\star$')
plt.ylabel(r'$\omega_k^\prime / m_\phi$')
plt.title(r'Frequência $\omega_k ^\prime / m_\phi , \xi = 11/60 , m_\chi/m_\phi = 1.5$')
#plt.xlim(1,10)
#plt.ylim(-1,40)
plt.legend()
plt.show()

```

## Resolvendo o sistema de equações diferenciais

$$\alpha_k '= -i\omega_k \alpha_k + \frac{\omega_k '}{2\omega_k} \beta_k$$

$$\beta_k '= i\omega_k \beta_k + \frac{\omega_k '}{2\omega_k} \alpha_k$$


### Caso $m_\chi = 7 \cdot 10^{11}$

`Comparar com figura 4 do artigo` [arXiv](https://arxiv.org/abs/1910.13937)

Neste caso, há a combinação entre o regime taquiônico e o de ressonâncias. Quando ocorre a transição entre os dois regimes, a frequência passa pelo zero, causando uma divergência(tendo em vista o modo que as equações foram escritas). Para resolver isso, retirei os pontos próximos da divergência, retornando derivadas nulas.  

```python
from scipy.integrate import odeint
from scipy.interpolate import interp1d
import scipy.integrate as integrate
import numpy as np
import matplotlib.pyplot as plt
import cmath

# Parâmetros constantes
k_values = np.logspace(-1, 2, num=70)  # Lista de valores de k
mx = 0.06  # Valor de mx
xi_values = [0.18, 0.26 , 0.43, 0.51]  # Valores de xi

# Funções para "a" e "R" dependentes do tempo
def a_function(t):
    return (t + 1)**2

def R_function(t):
    return (600 / (1.5335 + t)**6) * (0.5 * (1 - 3 * np.cos(0.37 * (t + 1.5335)**3)))

# Funções para as equações diferenciais
def derivative(y, t, k, xi):
    alpha_real, alpha_imag, beta_real, beta_imag = y

    a = a_function(t)
    R = R_function(t)

    omega_k = cmath.sqrt(k**2 + (a**2) * mx**2 + (a**2) * (xi - 1/6) * R)

    # Verificação da condição
    if omega_k.real + omega_k.imag < 0.003:
        return [0.0, 0.0, 0.0, 0.0]  # aqui eu pulo os pontos bem próximos onde a divergência ocorre, retornando derivadas nulas

    omega_kr = omega_k.real
    omega_ki = omega_k.imag

    omegap_k = 0.5 * (1 / omega_k) * (4 * (mx**2) * ((1 + t)**3) + 300 * (xi - 1/6) * (9 * np.sin(0.37 * (1.5335 + t)**3) * 0.37 * (((1 + t)**4) / (1.5335 + t)**4) - 2 * (1 - 3 * np.cos(0.37 * (1.5335 + t)**3)) * (((t * (1 + t)**3) / (1.5335 + t)**7))))

    deltar = (omegap_k / (2 * omega_k)).real

    alpha_real_prime = omega_kr * alpha_imag + omega_ki * alpha_real + deltar * beta_real
    alpha_imag_prime = -omega_kr * alpha_real + omega_ki * alpha_imag + deltar * beta_imag
    beta_real_prime = -omega_kr * beta_imag - omega_ki * beta_real + deltar * alpha_real
    beta_imag_prime = omega_kr * beta_real - omega_ki * beta_imag + deltar * alpha_imag

    return [alpha_real_prime, alpha_imag_prime, beta_real_prime, beta_imag_prime]

# Condições iniciais
alpha_real0 = 1.0
alpha_imag0 = 0.0
beta_real0 = 0.0
beta_imag0 = 0.0
initial_conditions = [alpha_real0, alpha_imag0, beta_real0, beta_imag0]

# Resolvendo o sistema de equações diferenciais para cada valor de xi
t_max = 20
num_points = 20000
time_points = np.linspace(0, t_max, num_points)

# Dicionário para armazenar os resultados para cada valor de xi
results_dict = {xi: {} for xi in xi_values}

for xi in xi_values:
    for k in k_values:
        result = odeint(derivative, initial_conditions, time_points, args=(k, xi))
        results_dict[xi][k] = result[:, 2:4]  # Armazenar ambas as partes real e imaginária de beta para este valor de k e xi

# Calcular as médias de beta_squared para cada valor de xi
beta_squared_averages = {xi: [] for xi in xi_values}

for xi in xi_values:
    for k in k_values:
        beta_squared = np.abs(results_dict[xi][k][-500:, 0] + 1j * (results_dict[xi][k][-500:, 1]))**2
        beta_squared_averages[xi].append((k**2) * np.mean(beta_squared))

# Plotar os resultados das médias de beta^2 em função de k para cada valor de xi
for xi in xi_values:
    plt.plot(k_values, beta_squared_averages[xi], '*', label=r'$\xi$=' + f'{xi}')

#plt.xlabel('Valor de k')
#plt.ylabel(r'Média de $|\beta_k|^2$')
#plt.title(r'Média de $|\beta_k|^2$ em função de k para diferentes valores de xi')
#plt.xscale("log")
#plt.legend()
#plt.grid()
#plt.show()

# Interpolação dos resultados para cada valor de xi
interp_functions = {}
for xi in xi_values:
    interp_functions[xi] = interp1d(k_values, beta_squared_averages[xi], kind='cubic')

# Plotar as funções interpoladas
k_values2 = np.logspace(-1, 2, num=500)
for xi in xi_values:
    plt.plot(k_values2, interp_functions[xi](k_values2), lw=2, label=r'$\xi$=' + f'{xi}')

plt.xlabel('k')
plt.ylabel(r'$k^2 |\beta_k|^2/m^2 _\phi $')
plt.title(r'Média de $|\beta_k|^2$ interpolada em função de k para diferentes valores de $\xi$')
plt.xscale("log")
plt.legend()
plt.grid()

# Calcular a integral das funções interpoladas para cada valor de xi
integral_results = {xi: integrate.simps(interp_functions[xi](k_values2)) for xi in xi_values}

# Imprimir os resultados das integrais
for xi in xi_values:
    print(f'Integral para xi={xi}: {integral_results[xi]}')
```

### Caso $m_\chi = 1.8 \cdot 10^{13}$

`Comparar com figura 3 do artigo` [arXiv](https://arxiv.org/abs/1910.13937)

Na figura do artigo, eles mostram um comportamento diferente do que encontramos aqui. Em nosso caso, ao aumentarmos o valor de $\xi$, aumentamos também produção a produção de partículas. No artigo, quanto maior o valor de $\xi$, menor a produção. A explicação que encontrei para isso é a seguinte: a condição inicial que eles escolheram engloba o regime inflacionário(deSitter), logo ao evoluir os passos de tempo no regime onde $m_\chi > m_\phi$, a massa efetiva do campo irá aumentar durante o regime inflacionário(caso aumentarmos o valor de $\xi$). Já em nosso código, o regime inflacionário "não existe", e ao aumentarmos o valor de $\xi$, as oscilações ficarão mais significantes e consequentemente a produção de partículas também irá aumentar.

```python
# Parâmetros constantes
k_values = np.logspace(-2, 1, num=50)  # Lista de valores de k
mx = 10**(-7)   # Valor de mx
xi_values = [0.38, 0.42 , 0.48]  # Valores de xi

# Funções para "a" e "R" dependentes do tempo
def a_function(t):
    return (t + 1)**2

def R_function(t):
    return (600 / (1.5335 + t)**6) * (0.5 * (1 - 3 * np.cos(0.37 * (t + 1.5335)**3)))

# Funções para as equações diferenciais
def derivative(y, t, k, xi):
    alpha_real, alpha_imag, beta_real, beta_imag = y

    a = a_function(t)
    R = R_function(t)

    omega_k = cmath.sqrt(k**2 + (a**2) * mx**2 + (a**2) * (xi - 1/6) * R)

    # Verificação da condição
    if omega_k.real + omega_k.imag < 0.2:
        return [0.0, 0.0, 0.0, 0.0]  # Retorna derivadas zero para pular os pontos

    omega_kr = omega_k.real
    omega_ki = omega_k.imag

    omegap_k = 0.5 * (1 / omega_k) * (4 * (mx**2) * ((1 + t)**3) + 300 * (xi - 1/6) * (9 * np.sin(0.37 * (1.5335 + t)**3) * 0.37 * (((1 + t)**4) / (1.5335 + t)**4) - 2 * (1 - 3 * np.cos(0.37 * (1.5335 + t)**3)) * (((t * (1 + t)**3) / (1.5335 + t)**7))))

    deltar = (omegap_k / (2 * omega_k)).real

    alpha_real_prime = omega_kr * alpha_imag + omega_ki * alpha_real + deltar * beta_real
    alpha_imag_prime = -omega_kr * alpha_real + omega_ki * alpha_imag + deltar * beta_imag
    beta_real_prime = -omega_kr * beta_imag - omega_ki * beta_real + deltar * alpha_real
    beta_imag_prime = omega_kr * beta_real - omega_ki * beta_imag + deltar * alpha_imag

    return [alpha_real_prime, alpha_imag_prime, beta_real_prime, beta_imag_prime]

# Condições iniciais
alpha_real0 = 1.0
alpha_imag0 = 0.0
beta_real0 = 0.0
beta_imag0 = 0.0
initial_conditions = [alpha_real0, alpha_imag0, beta_real0, beta_imag0]

# Resolvendo o sistema de equações diferenciais para cada valor de xi
t_max = 20
num_points = 20000
time_points = np.linspace(0, t_max, num_points)

# Dicionário para armazenar os resultados para cada valor de xi
results_dict = {xi: {} for xi in xi_values}

for xi in xi_values:
    for k in k_values:
        result = odeint(derivative, initial_conditions, time_points, args=(k, xi))
        results_dict[xi][k] = result[:, 2:4]  # Armazenar ambas as partes real e imaginária de beta para este valor de k e xi

# Calcular as médias de beta_squared para cada valor de xi
beta_squared_averages = {xi: [] for xi in xi_values}

for xi in xi_values:
    for k in k_values:
        beta_squared = np.abs(results_dict[xi][k][-500:, 0] + 1j * (results_dict[xi][k][-500:, 1]))**2
        beta_squared_averages[xi].append((k**2) * np.mean(beta_squared))

# Plotar os resultados das médias de beta^2 em função de k para cada valor de xi
for xi in xi_values:
    plt.plot(k_values, beta_squared_averages[xi], '*', label=r'$\xi$=' + f'{xi}')

plt.xlabel('Valor de k')
plt.ylabel(r'Média de $|\beta_k|^2$')
plt.title(r'Média de $|\beta_k|^2$ em função de k para diferentes valores de xi')
plt.xscale("log")
plt.legend()
plt.grid()
plt.show()

# Interpolação dos resultados para cada valor de xi
interp_functions = {}
for xi in xi_values:
    interp_functions[xi] = interp1d(k_values, beta_squared_averages[xi], kind='cubic')

# Plotar as funções interpoladas
k_values2 = np.logspace(-2, 1, num=500)
for xi in xi_values:
    plt.plot(k_values2, interp_functions[xi](k_values2), lw=2, label=r'$\xi$=' + f'{xi}')

plt.xlabel('k')
plt.ylabel(r'$k^2 |\beta_k|^2/m^2 _\phi $')
plt.title(r'Média de $|\beta_k|^2$ interpolada em função de k para diferentes valores de $\xi$')
plt.xscale("log")
plt.legend()
plt.grid()

# Calcular a integral das funções interpoladas para cada valor de xi
integral_results = {xi: integrate.simps(interp_functions[xi](k_values2)) for xi in xi_values}

# Imprimir os resultados das integrais
for xi in xi_values:
    print(f'Integral para xi={xi}: {integral_results[xi]}')
```
